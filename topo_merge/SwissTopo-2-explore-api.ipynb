## STAC

```python
import json
from urllib.parse import urlencode

def apiget(url, **params):
    if params:
        if not url.endswith('?'): url += '?'
        url += urlencode(params) #, quote_via=quote_plus)
    print('GET', url)
    return json.loads(urlopen(url).read())

def get_first_link(data, rel):
    return [lnk for lnk in data['links'] if lnk['rel'] == rel][0]['href']
```

```python
# same with STAC (data API)
collections = apiget("https://data.geo.admin.ch/api/stac/v0.9/collections")
```

```python
[c['id'] for c in collections['collections'] if 'ch.swisstopo' in c['id']]
```

```python
data_desc = [c for c in collections['collections'] if c['id'] == 'ch.swisstopo.pixelkarte-farbe-pk200.noscale'][0]
data_desc
```

```python
items = apiget(get_first_link(data_desc, 'items'))
print('\n'.join(f['id'] for f in items['features']))
```

```python
len(items['features'])
```

```python
data10_desc = [c for c in collections['collections'] if c['id'] == 'ch.swisstopo.pixelkarte-farbe-pk25.noscale'][0]
items10 = apiget(get_first_link(data10_desc, 'items'))
print(len(items10['features']))
```
